```python
import numpy as np
import pandas as pd
```

## Task 1: Pandas â†’ NumPy (Bridge Task)

```python
df = pd.read_csv('car-sales-missing-data.csv')
df
```

```python
df['Price']=df['Price'].str.replace('$', '', regex = False)
df['Price']=df['Price'].str.replace(",", "", regex = False) 
df['Price']= df['Price'].astype(float)
df.info()
```

```python
df['Odometer'].fillna(df['Odometer'].median(), inplace = True)
df['Doors'].fillna(df['Doors'].mode()[0], inplace = True)
df['Price'].fillna(df['Price'].median(), inplace = True)
```

```python
df.info()
```

```python
np_arr = df.select_dtypes(include = ['float', 'int']).to_numpy(dtype='int')
np_arr
```

## Task 4: Column-wise Scaling (Manual)

```python
# Scale the Odometer column using this formula:
scaled_odo= (np_arr[:,0] - np.mean(np_arr[:,0]))/np.std(np_arr[:,0]) 
scaled_odo
```

## Task 5: Back to Pandas (ML Thinking)

```python
# Take the scaled Odometer values and: Replace the original Odometer column in df
df["Odometer"] = scaled_odo
df.head()
```
