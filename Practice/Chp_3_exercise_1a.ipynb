## An MNIST Classifier With Over 97% Accuracy   
Exercise: Try to build a classifier for the MNIST dataset that achieves over 97% accuracy on the test set. Hint: the KNeighborsClassifier works quite well for this task; you just need to find good hyperparameter values (try a grid search on the weights and n_neighbors hyperparameters).

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
mnist = fetch_openml("mnist_784", as_frame=False)
```

```python
print(mnist.DESCR)
```

```python
mnist.keys()
```

```python
X, y = mnist.data, mnist.target
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state= 42)
```

```python
knn_1 = KNeighborsClassifier()
knn_1.fit(X_train, y_train)
```

```python
knn_1_accuracy = knn_1.score(X_test, y_test)
```

```python
knn_1_accuracy
```

```python
knn_1.get_params()
```

```python
param_grid = [
    {'weights' : ['uniform', 'distance'], 
    'n_neighbors': [3, 4, 5, 6]}]

knn_clf_2= KNeighborsClassifier()
grid_seacrch = GridSearchCV(knn_clf_2, param_grid, cv=5)
grid_seacrch.fit(X_train[:10_000], y_train[:10_000])
```

```python
knn_2_grid_cv_accuracy = grid_seacrch.score(X_test, y_test)
```

```python
knn_2_grid_cv_accuracy
```

```python
grid_seacrch.best_estimator_.fit(X_train, y_train)
```

```python
tuned_accuracy = grid_seacrch.score(X_test, y_test)
tuned_accuracy
```
