# Classification Fundamentals, MNIST, Binary Classification

## Import Libraries

````python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
````

## Load MNIST Dataset

````python
mnist = fetch_openml('mnist_784', version = 1, as_frame= False)

X = mnist.data
y = mnist.target.astype(int)
````

## Convert to binary classification

````python
y_binary = (y==5)  # True if 5, False otherwise
````

## Train Test Split

````python
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
````

## Train a simple classifier

````python
clf = SGDClassifier(random_state=42)

clf.fit(X_train, y_train)
````

## Predctions

````python
y_pred = clf.predict(X_test)
````

## Accuracy

````python
acc = accuracy_score(y_test, y_pred)
print("Accuracy : ", acc)
````

## Confusion Matrix

````python
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix : ")
print(cm)
````

```python 
[[TN  FP] 
[FN  TP]] 


Row 0 → actual class False / 0 (Negative)

Row 1 → actual class True / 1 (Positive)

Column 0 → predicted Negative

Column 1 → predicted Positive
```

````python
TN, FP, FN, TP = cm.ravel()

print("True Negatives (not 5 correctly) :", TN)
print("Flase Positives (not 5 predicted as 5):", FP)
print("False Negatives (5 missed)", FN)
print("True Negatives (5 correctly detected)", TP)
````

## Intentional BAD model (accuracy trap)

````python
bad_preds = np.zeros_like(y_test, dtype =bool)   # always predicts "not 5"

print("Bad Model Accuracy :", accuracy_score(y_test, bad_preds) )
print("Bad Model Confusion Matrix:")
print(confusion_matrix(y_test, bad_preds))
````
